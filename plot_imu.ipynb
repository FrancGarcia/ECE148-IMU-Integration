```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your data
df = pd.read_csv("imu_data.csv")  # Replace with your file path

# Define a moving average function
def moving_average(data, window_size):
    return data.rolling(window=window_size, center=True).mean()

# Apply the moving average to accelerometer data
window_size = 10  # Adjust as needed
df["aX_smooth"] = moving_average(df["aX"], window_size)
df["aY_smooth"] = moving_average(df["aY"], window_size)
df["aZ_smooth"] = moving_average(df["aZ"], window_size)
# Plot the results
plt.figure(figsize=(12, 6))
plt.plot(df["aX"], alpha=0.5, label="Raw aX", color="blue")
plt.plot(df["aX_smooth"], label="Smoothed aX", color="red")
plt.plot(df["aY"], alpha=0.5, label="Raw aY", color="orange")
plt.plot(df["aY_smooth"], label="Smoothed aY", color="green")
plt.legend()
plt.title("Smoothed Accelerometer Data")
plt.show()
```

```python
from scipy.signal import savgol_filter

df["aX_smooth"] = savgol_filter(df["aX"], window_length=11, polyorder=2)
df["aY_smooth"] = savgol_filter(df["aY"], window_length=11, polyorder=2)

plt.figure(figsize=(12, 6))
plt.plot(df["aX"], alpha=0.5, label="Raw aX", color="blue")
plt.plot(df["aX_smooth"], label="Smoothed aX", color="red")
plt.plot(df["aY"], alpha=0.5, label="Raw aY", color="orange")
plt.plot(df["aY_smooth"], label="Smoothed aY", color="green")
plt.legend()
plt.title("Smoothed Accelerometer Data")
plt.show()

```
